```python
import pandas as pd
from sklearn.metrics import mean_squared_error

# Carica il file CSV
df = pd.read_csv("best_overall_results.csv")

# Estrai le colonne
x_pred = df["x_pred"]
y_pred = df["y_pred"]
x_real = df["x_real"]
y_real = df["y_real"]

# MSE separati
mse_x = mean_squared_error(x_real, x_pred)
mse_y = mean_squared_error(y_real, y_pred)

# MSE complessivo su entrambe le coordinate
mse_xy = mean_squared_error(df[["x_real", "y_real"]], df[["x_pred", "y_pred"]])

print("MSE X:", mse_x)
print("MSE Y:", mse_y)
print("MSE totale (x,y):", mse_xy)

# Conta quante volte l'errore assoluto supera 0.33
count_over_033 = ((abs(x_pred - x_real) > 0.33) | (abs(y_pred - y_real) > 0.33)).sum()
count_over_055 = ((abs(x_pred - x_real) > 0.55) | (abs(y_pred - y_real) > 0.55)).sum()

print("Numero di predizioni con errore > 0.33:", count_over_033)
print("Numero di predizioni con errore > 0.55:", count_over_055)
```
